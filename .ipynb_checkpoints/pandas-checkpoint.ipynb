```python
import pandas as pd
df1 = pd.DataFrame({'name':['Ram','Shyam','Hari'],
       'age':[45,67,89],
       'add':['Kathmandu','Bhaktapur','Lalitpur']})
df2 = pd.DataFrame({'name':['Rama','Shyama','Hary'],
       'age':[38,56,98],
       'add':['Tinkune','Koteswor','Baneswor']})
df3 = pd.DataFrame({'name':['John','Sam','Jack'],
       'age':[32,32,31],
       'add':['Kathmandu','Bhaktapur','Lalitpur']})

df = pd.concat([df1,df2,df3])
df
```
